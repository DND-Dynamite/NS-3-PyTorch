# Generate Auth Model

```python
#!/usr/bin/env python3
"""
train_auth_model.py
Physical-layer authentication via Neyman-Pearson hypothesis test
using distance-dependent path-loss as device fingerprint.

Input : CSV files produced by generate_thz_data.py
Output: ROC curve, detection/miss probabilities vs threshold,
        feasible-region illustration (ξ vs |d_AB - d_EB|)
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.special import erfc
from pathlib import Path

# --------------------------------------------------
# Configuration
# --------------------------------------------------
data_dir = Path("data")

# System parameters from the paper
f_THz = 1.2          # centre frequency [THz]
N = 4                # number of sweat ducts
d_AB_mm = 15.0       # Alice–Bob distance

# --------------------------------------------------
# Utilities
# --------------------------------------------------
def Q(x):
    """Gaussian Q-function = 0.5*erfc(x/sqrt(2))"""
    return 0.5 * erfc(x / np.sqrt(2))

def Qinv(q):
    """Inverse Q-function"""
    from scipy.special import ndtri
    return -ndtri(q)   # Corrected: Q^{-1}(q) = -Φ^{-1}(q)

def pathloss(d_mm, f_THz, N):
    """Path-loss model from paper eq-1"""
    return (-0.2*N + 3.98) \
           + (0.44*N + 98.48)*np.power(d_mm, 0.65) \
           + (0.068*N + 2.4)*np.power(f_THz, 4.07)

# --------------------------------------------------
# 1. Load data
# --------------------------------------------------
try:
    train = pd.read_csv(data_dir/"alice_train.csv")
    alice = pd.read_csv(data_dir/"alice_test.csv")
    eve = pd.read_csv(data_dir/"eve_test.csv")
    
    pl_AB = train.true_pl.iloc[0]          # ground truth
    sigma = alice.meas_pl.std()            # estimated noise std
    
    print(f"Loaded data successfully:")
    print(f"  Training samples: {len(train)}")
    print(f"  Alice test samples: {len(alice)}")
    print(f"  Eve test samples: {len(eve)}")
    print(f"  Ground truth PL_AB: {pl_AB:.2f} dB")
    print(f"  Estimated noise std: {sigma:.2f} dB")
    
except FileNotFoundError as e:
    print(f"Error: {e}")
    print("Please run generate_thz_data.py first to create the data files.")
    exit(1)

# --------------------------------------------------
# 2. Neyman-Pearson threshold for chosen Pfa
# --------------------------------------------------
Pfa_set = 1e-3
delta = sigma * Qinv(Pfa_set / 2)
print(f"\nNeyman-Pearson threshold:")
print(f"  Chosen Pfa = {Pfa_set}  →  δ = {delta:.2f} dB")

# --------------------------------------------------
# 3. Evaluate on test sets
# --------------------------------------------------
def decide(pl_meas, ground, thr):
    """Return 0 (accept) if |pl_meas - ground| <= thr else 1 (reject)"""
    return np.where(np.abs(pl_meas - ground) <= thr, 0, 1)

alice_pred = decide(alice.meas_pl, pl_AB, delta)
eve_pred = decide(eve.meas_pl, pl_AB, delta)

Pfa_real = (alice_pred == 1).mean()
Pmd = (eve_pred == 0).mean()
Pd = 1 - Pmd

print(f"\nAuthentication Performance:")
print(f"  Empirical Pfa = {Pfa_real:.4f}")
print(f"  Empirical Pmd = {Pmd:.4f}")
print(f"  Detection rate Pd = {Pd:.4f}")

# --------------------------------------------------
# 4. ROC curve (sweep delta)
# --------------------------------------------------
print(f"\nGenerating ROC curve...")
thr_range = np.linspace(0, 4*sigma, 200)
pfa_roc, pmd_roc = [], []
for thr in thr_range:
    pfa_roc.append((decide(alice.meas_pl, pl_AB, thr) == 1).mean())
    pmd_roc.append((decide(eve.meas_pl, pl_AB, thr) == 0).mean())

pfa_roc, pmd_roc = np.array(pfa_roc), np.array(pmd_roc)
pd_roc = 1 - pmd_roc

plt.figure(figsize=(6, 5))
plt.plot(pfa_roc, pd_roc, 'b-', linewidth=2, label='ROC Curve')
plt.plot([0, 1], [0, 1], 'k--', lw=1, alpha=0.7, label='Random Guess')
plt.scatter([Pfa_real], [Pd], color='red', s=80, zorder=5, 
           label=f'Operating Point (Pfa={Pfa_real:.3f}, Pd={Pd:.3f})')
plt.xlabel("Probability of False Alarm (Pfa)")
plt.ylabel("Probability of Detection (Pd)")
plt.title("ROC Curve - Path-loss Authentication")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig(data_dir/"roc_curve.png", dpi=150, bbox_inches='tight')
plt.show()

# --------------------------------------------------
# 5. Deflection coefficient vs |d_AB - d_EB|
# --------------------------------------------------
print(f"\nGenerating feasible region analysis...")

def deflection_coeff(d_diff_mm, sigma):
    """
    ξ = (E[T|H1]-E[T|H0])^2 / Var[T|H0]
    Closed-form for folded-normal statistic T=|z-pl_AB| under
    H0: z ~ N(pl_AB, σ²)
    H1: z ~ N(pl_EB, σ²) where Δ = pl_EB - pl_AB
    See paper eq-8,9.
    """
    # Calculate actual path loss difference using the model
    pl_AB = pathloss(d_AB_mm, f_THz, N)
    pl_EB = pathloss(d_AB_mm + d_diff_mm, f_THz, N)
    Delta = pl_EB - pl_AB
    
    # Expected value under H0
    ET_H0 = sigma * np.sqrt(2 / np.pi)
    
    # Variance under H0
    Var_H0 = sigma**2 * (1 - 2 / np.pi)
    
    # Expected value under H1 (folded normal)
    tmp = Delta / sigma
    from scipy.stats import norm
    ET_H1 = sigma * np.sqrt(2 / np.pi) * np.exp(-0.5 * tmp**2) + Delta * (1 - 2 * norm.cdf(-Delta / sigma))
    
    xi = (ET_H1 - ET_H0)**2 / Var_H0
    return xi

# Calculate Eve's actual distance from the data
pl_EB_data = eve.true_pl.iloc[0]
d_EB_actual = d_AB_mm + 2.0  # Approximate based on known setup from data generation
actual_d_diff = abs(d_AB_mm - d_EB_actual)

d_diffs = np.linspace(0, 5, 100)
xis = [deflection_coeff(d, sigma) for d in d_diffs]

plt.figure(figsize=(7, 5))
plt.plot(d_diffs, xis, 'b-', linewidth=2, label='Deflection Coefficient')

# Mark the actual operating point
xi_actual = deflection_coeff(actual_d_diff, sigma)
plt.axvline(actual_d_diff, ls='--', color='red', alpha=0.7, 
           label=f'Actual |d_AB-d_EB| = {actual_d_diff:.1f} mm')
plt.scatter([actual_d_diff], [xi_actual], color='red', s=80, zorder=5)

# Paper reference values
plt.axhline(166.7, ls='--', color='gray', alpha=0.7, label=r'Paper $\beta$=166.7')
plt.axvline(2.0, ls='--', color='gray', alpha=0.7, label=r'Paper $\eta$=2.0 mm')

# Shade feasible region
feasible_mask = np.array(xis) >= 166.7
if np.any(feasible_mask):
    feasible_start = d_diffs[feasible_mask][0]
    plt.axvspan(feasible_start, max(d_diffs), alpha=0.2, color='green', label='Feasible Region')

plt.xlabel(r'Distance Difference $|d_{AB}-d_{EB}|$ [mm]')
plt.ylabel(r'Deflection Coefficient $\xi$')
plt.title("Feasible Region Analysis")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig(data_dir/"feasible_region.png", dpi=150, bbox_inches='tight')
plt.show()

print(f"\nFeasible Region Analysis:")
print(f"  Actual distance difference: {actual_d_diff:.1f} mm")
print(f"  Actual deflection coefficient: {xi_actual:.1f}")
print(f"  Paper threshold (β): 166.7")
print(f"  Authentication feasible: {'YES' if xi_actual >= 166.7 else 'NO'}")

# --------------------------------------------------
# 6. Additional: Pfa and Pd vs Threshold
# --------------------------------------------------
plt.figure(figsize=(8, 5))

plt.subplot(1, 2, 1)
plt.plot(thr_range, pfa_roc, 'r-', linewidth=2, label='False Alarm (Pfa)')
plt.plot(thr_range, pmd_roc, 'b-', linewidth=2, label='Miss Detection (Pmd)')
plt.axvline(delta, color='black', linestyle='--', label=f'NP Threshold (δ={delta:.1f} dB)')
plt.xlabel('Threshold δ [dB]')
plt.ylabel('Probability')
plt.title('Pfa and Pmd vs Threshold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(thr_range, pd_roc, 'g-', linewidth=2, label='Detection (Pd)')
plt.axvline(delta, color='black', linestyle='--', label=f'NP Threshold (δ={delta:.1f} dB)')
plt.xlabel('Threshold δ [dB]')
plt.ylabel('Probability of Detection')
plt.title('Pd vs Threshold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(data_dir/"threshold_analysis.png", dpi=150, bbox_inches='tight')
plt.show()

print(f"\nAnalysis complete! Generated plots:")
print(f"  - roc_curve.png")
print(f"  - feasible_region.png") 
print(f"  - threshold_analysis.png")
```
